```python
import matplotlib.pyplot as plt

# Dữ liệu mẫu
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('My Plot')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'city': ['NYC', 'LA', 'Chicago']
})
```

```python
# Dùng dataframe df đã tạo từ trước
plt.scatter(df['age'], df['city'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot')
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')
plt.show()
```

```python
import seaborn as sns

df_numeric = pd.DataFrame({
    'age': [25, 30, 35, 28, 40],
    'salary': [50000, 60000, 80000, 55000, 90000],
    'experience': [2, 5, 10, 3, 15]
})

# Dùng dataframe df
sns.heatmap(df_numeric.corr(), annot=True)  # Hiển thị ma trận tương quan
plt.show()
```
