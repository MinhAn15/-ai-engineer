```python
# Linear Regression đơn giản: y = mx + b
# m: slope (hệ số góc)
# b: intercept (hệ số chặn)

# Ví dụ: dự đoán giá nhà từ diện tích
# y (giá nhà) = m * x (diện tích) + b
```

```python
# y = w₁x₁ + w₂x₂ + w₃x₃ + ... + wₙxₙ + b

# Ví dụ: giá nhà từ nhiều yếu tố
# y (giá) = w₁*(diện tích) + w₂*(số phòng) + w₃*(tuổi nhà) + ... + b

# w₁, w₂, w₃, ... : weights (trọng số)
# b: bias (hệ số chặn)
```

```python
# Fitting = tìm w và b tốt nhất

# "Tốt nhất" = minimize sai số giữa y_thật và y_dự_đoán

# Thuật toán: Ordinary Least Squares (OLS)
# Tìm w, b sao cho tổng bình phương sai số nhỏ nhất
```

```python
# y_i: giá trị thật
# ŷ_i: giá trị dự đoán
# n: số samples

# MSE đo "trung bình bình phương sai số"
# Càng nhỏ = model càng chính xác
```

```python
# R² = 1.0: model dự đoán hoàn hảo
# R² = 0.0: model ngang với dự đoán bằng mean(y)
# R² < 0.0: model tệ hơn cả việc dự đoán mean
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo data mẫu
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.randn(50) * 2

# Vẽ data points
plt.scatter(x, y, alpha=0.5)

# Vẽ regression line (giả sử w=2, b=1)
x_line = np.array([0, 10])
y_line = 2 * x_line + 1
plt.plot(x_line, y_line, 'r-', label='y = 2x + 1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression: Fitting a Line')
plt.legend()
plt.show()
```
